```python
import numpy as np
import matplotlib.pyplot as plt

Uk = np.arange(16)

bias_values = [0, -3, 2]

for bk in bias_values:
    Vk = Uk + bk
    plt.plot(Uk, Vk, label=f'bk = {bk}')

plt.xlabel('Uk')
plt.ylabel('Vk')
plt.legend()
plt.title('Effect of Bias (bk) on Vk vs. Uk')
plt.grid(True)

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display an image
def show_image(image, title):
    plt.figure(figsize=(8, 8))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
input_image = cv2.imread('/content/Fig0236(a)(letter_T) (1).tif')
```

```python
identity_image = input_image.copy()
show_image(input_image, 'Original Image')
```

```python
scale_factor = 0.5
scaled_image = cv2.resize(input_image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_LINEAR)

show_image(scaled_image, 'Scaled')
```

```python
angle_degrees = 45
height, width = input_image.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle_degrees, 1)
rotated_image = cv2.warpAffine(input_image, rotation_matrix, (width, height))
show_image(rotated_image, 'Rotated')

```

```python

tx, ty = 50, 30
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
translated_image = cv2.warpAffine(input_image, translation_matrix, (width, height))
show_image(translated_image, 'Translated')
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt


def vertical_shear(image, shear_factor):
    rows, cols, _ = image.shape
    M = np.array([[1, shear_factor, 0], [0, 1, 0]], dtype=np.float32)
    sheared_image = cv2.warpAffine(image, M, (cols, rows))
    return sheared_image

def horizontal_shear(image, shear_factor):
    rows, cols, _ = image.shape
    M = np.array([[1, 0, 0], [shear_factor, 1, 0]], dtype=np.float32)
    sheared_image = cv2.warpAffine(image, M, (cols, rows))
    return sheared_image




```

```python
image_path =('/content/Fig0236(a)(letter_T) (1).tif')
original_image = cv2.imread(image_path)
```

```python
vertical_shear_factor = 0.2
horizontal_shear_factor = 0.2

vertical_sheared_image = vertical_shear(original_image, vertical_shear_factor)
horizontal_sheared_image = horizontal_shear(original_image, horizontal_shear_factor)

plt.figure(figsize=(10, 6))

plt.subplot(131)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(132)
plt.title('Vertical Shear')
plt.imshow(cv2.cvtColor(vertical_sheared_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(133)
plt.title('Horizontal Shear')
plt.imshow(cv2.cvtColor(horizontal_sheared_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.tight_layout()
plt.show()

```
